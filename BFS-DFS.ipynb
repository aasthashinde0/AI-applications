```python
from collections import deque

# Capacity of jugs
JUG1 = 4
JUG2 = 3
GOAL = 2

# To keep track of visited states
visited = set()

# ---------- BFS Implementation ----------
def bfs():
    print("BFS Solution:")
    visited.clear()
    queue = deque()
    queue.append(((0, 0), []))  # (state, path)

    while queue:
        (a, b), path = queue.popleft()
        print(f"Current State: ({a}, {b})")

        if (a, b) in visited:
            continue
        visited.add((a, b))

        path = path + [(a, b)]

        if a == GOAL or b == GOAL:
            print(f"\nReached target:({a}, {b})")
            print("\nPath to solution:")
            for step in path:
                print(f"{step[0]} , {step[1]}")
            return
           

        # Generate all possible next states
        next_states = get_next_states(a, b)
        for state in next_states:
            if state not in visited:
                queue.append((state, path))
    print("No solution found.")

# ---------- DFS Implementation ----------

def dfs(a, b, path):    
    print(f"Current State: ({a}, {b})")
    
    if a == GOAL or b == GOAL:
        print(f"\nReached target: ({a}, {b})")
        print("Steps:")
        for step in path + [(a, b)]:
            print(f"{step[0]} , {step[1]}")
        return True

    if (a, b) in visited:
        return False

    visited.add((a, b))

    for next_a, next_b in get_next_states(a, b):
        if dfs(next_a, next_b, path + [(a, b)]):
            return True

    return False

# ---------- Generate all valid next states ----------

def get_next_states(a, b):
    states = []
    # Fill Jug1
    states.append((JUG1, b))
    # Fill Jug2
    states.append((a, JUG2))
    # Empty Jug1
    states.append((0, b))
    # Empty Jug2
    states.append((a, 0))
    # Pour Jug1 → Jug2
    pour = min(a, JUG2 - b)
    states.append((a - pour, b + pour))
    # Pour Jug2 → Jug1
    pour = min(b, JUG1 - a)
    states.append((a + pour, b - pour))
    return states

# ---------- Main ----------
bfs()
print(f"\nDFS Solution:")
visited.clear()
dfs(0, 0, [])
```
